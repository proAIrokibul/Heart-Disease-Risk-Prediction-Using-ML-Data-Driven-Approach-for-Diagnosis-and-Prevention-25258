### <i> Data Preprocessing </i>

```python
# Basic libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 2. Remove duplicates (if any)
df.drop_duplicates(inplace=True)
print(f"✅ Data shape after removing duplicates: {df.shape}")

# 3. Check for missing values
missing = df.isnull().sum()
print("\n🔍 Missing values in dataset:")
print(missing[missing > 0] if missing.any() else "No missing values found.")

# 4. Encode categorical features
# Gender is likely binary: 0 = Female, 1 = Male (already encoded as float)
df['Gender'] = df['Gender'].astype(int)

# Optional: convert Heart_Risk to int as well
df['Heart_Risk'] = df['Heart_Risk'].astype(int)

# 5. Define features and target
X = df.drop('Heart_Risk', axis=1)
y = df['Heart_Risk']

# 6. Feature Scaling using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 7. Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# 8. Final confirmation
print(f"\n📐 Final dataset shapes:")
print(f"X_train: {X_train.shape}")
print(f"X_test : {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test : {y_test.shape}")
```

### <i> Modeling </i>
### <i> Logistic Regerssion </i>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score

# Pipeline and GridSearch
lr_pipeline = Pipeline([
    ('classifier', LogisticRegression(solver='liblinear', max_iter=1000))
])

lr_params = {
    'classifier__C': [0.01, 0.1, 1, 10],
    'classifier__penalty': ['l2']
}

lr_grid = GridSearchCV(lr_pipeline, lr_params, cv=5, scoring='accuracy', n_jobs=-1)
lr_grid.fit(X_train, y_train)

# Results
print("🔍 Best Logistic Regression Parameters:")
print(lr_grid.best_params_)

y_pred_lr = lr_grid.predict(X_test)
print("\n📊 Classification Report (Logistic Regression):")
print(classification_report(y_test, y_pred_lr))
print("✅ Accuracy:", accuracy_score(y_test, y_pred_lr))
```

### <i> Random Forest Classifier </i>

```python
from sklearn.ensemble import RandomForestClassifier

# Pipeline and GridSearch
rf_pipeline = Pipeline([
    ('classifier', RandomForestClassifier(random_state=42))
])

rf_params = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [5, 10, 20],
    'classifier__min_samples_split': [2, 5],
    'classifier__min_samples_leaf': [1, 2]
}

rf_grid = GridSearchCV(rf_pipeline, rf_params, cv=5, scoring='accuracy', n_jobs=-1)
rf_grid.fit(X_train, y_train)

# Results
print("🔍 Best Random Forest Parameters:")
print(rf_grid.best_params_)

y_pred_rf = rf_grid.predict(X_test)
print("\n📊 Classification Report (Random Forest):")
print(classification_report(y_test, y_pred_rf))
print("✅ Accuracy:", accuracy_score(y_test, y_pred_rf))
```

### <i> XGB Classifier </i>

```python
from xgboost import XGBClassifier

# Pipeline and GridSearch
xgb_pipeline = Pipeline([
    ('classifier', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42))
])

xgb_params = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [3, 5, 7],
    'classifier__learning_rate': [0.01, 0.1, 0.2],
    'classifier__subsample': [0.8, 1.0]
}

xgb_grid = GridSearchCV(xgb_pipeline, xgb_params, cv=5, scoring='accuracy', n_jobs=-1)
xgb_grid.fit(X_train, y_train)

# Results
print("🔍 Best XGBoost Parameters:")
print(xgb_grid.best_params_)

y_pred_xgb = xgb_grid.predict(X_test)
print("\n📊 Classification Report (XGBoost):")
print(classification_report(y_test, y_pred_xgb))
print("✅ Accuracy:", accuracy_score(y_test, y_pred_xgb))
```

### <i> Comparision Of All Models </i>

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix

def plot_model_results(model_name, y_train, y_train_pred, y_test, y_test_pred):
    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc = accuracy_score(y_test, y_test_pred)

    print(f"📌 {model_name} | Train Accuracy: {train_acc:.4f}, Test Accuracy: {test_acc:.4f}")

    # 1. Accuracy Bar Plot
    plt.figure(figsize=(6, 4))
    sns.barplot(x=['Train Accuracy', 'Test Accuracy'], y=[train_acc, test_acc], palette='Blues')
    plt.title(f'{model_name} Accuracy Comparison')
    plt.ylim(0, 1)
    plt.ylabel('Accuracy')
    plt.show()

    # 2. Confusion Matrix (Test Set)
    cm = confusion_matrix(y_test, y_test_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges')
    plt.title(f'{model_name} Confusion Matrix (Test Set)')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # 3. Distribution Plot: Actual vs Predicted
    plt.figure(figsize=(6, 4))
    sns.histplot(y_test, label='Actual', color='black', bins=2, alpha=0.6, stat="density")
    sns.histplot(y_test_pred, label='Predicted', color='orange', bins=2, alpha=0.6, stat="density")
    plt.legend()
    plt.title(f'{model_name} | Actual vs Predicted (Test Set)')
    plt.xlabel('Class')
    plt.show()
```

```python
# LOGISTIC REGRESSION
plot_model_results(
    "Logistic Regression",
    y_train, lr_grid.predict(X_train),
    y_test, lr_grid.predict(X_test)
)

```

```python
# RANDOM FOREST
plot_model_results(
    "Random Forest",
    y_train, rf_grid.predict(X_train),
    y_test, rf_grid.predict(X_test)
)


```

```python
# XGBOOST
plot_model_results(
    "XGBoost",
    y_train, xgb_grid.predict(X_train),
    y_test, xgb_grid.predict(X_test)
)
```
